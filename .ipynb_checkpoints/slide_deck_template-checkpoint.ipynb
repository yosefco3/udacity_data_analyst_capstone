## Investigation Overview

> Can we identify malignancy by parameters like shape and size ?

## Dataset Overview

> The dataset is from kaggle.     
> Features in the data set are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("data.csv")
df.drop(columns=["Unnamed: 32"],inplace=True)
df.head(2)

```

## Diagnosis

> How many malingnant and benign we have in the dataset ?


```python
diag = df["diagnosis"].value_counts()
fig, ax = plt.subplots(1, 1,figsize = (10,8))
sns.barplot(x=diag.index, y=diag.values,ax=ax)
# plt.bar(diag.index,diag.values, align='center');
# plt.xticks(y_pos, objects)
a = plt.ylabel('Number of cases')
b = plt.xlabel('M\B')
c = plt.title('malignant Vs benign')
```

### The most cases in the dataset (above 60%) are benign.

## Connection between features and malingnancy:

## radius_mean and diagnosis
    * we could see here overlapping between B & M.
    * we could see that the overlapping is only on part of the range.
    * we could see that the min,max,IQR, and the average are different between two plots.
    * we could see outliers.

```python
columns =  ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']
df_small = df[columns+['id','diagnosis']]
df_small.set_index('id',inplace=True)
# df_small.head(2)
```

```python

fig, (ax1,ax2) = plt.subplots(1, 2,figsize = (23,8))
pal = ['orange', 'blue']
sns.kdeplot(ax=ax1, data=df_small, x='radius_mean', hue="diagnosis",fill=True,palette=pal);
sns.boxplot(ax=ax2, x="diagnosis", y='radius_mean', data=df_small,palette=pal);
a = ax1.set(title=f"{'radius_mean'} overlapping malingnant and benign",xlabel=f"{'radius_mean'}",ylabel="count")
b = ax2.set(title=f"{'radius_mean'} overlapping malingnant and benign",xlabel=f"{'radius_mean'}",ylabel="count")
```

## compactness_mean and diagnosis:
    * we could see here overlapping between B & M.
    * we could see that the overlapping is only on part of the range.
    * we could see that the min,max,IQR, and the average are different between two plots.
    * we could see outliers.

```python

fig, (ax1,ax2) = plt.subplots(1, 2,figsize = (23,8))
pal = ['orange', 'blue']
sns.kdeplot(ax=ax1, data=df_small, x='compactness_mean', hue="diagnosis",fill=True,palette=pal);
sns.boxplot(ax=ax2, x="diagnosis", y='compactness_mean', data=df_small,palette=pal);
a = ax1.set(title=f"{'compactness_mean'} overlapping malingnant and benign",xlabel=f"{'compactness_mean'}",ylabel="count")
b = ax2.set(title=f"{'compactness_mean'} overlapping malingnant and benign",xlabel=f"{'compactness_mean'}",ylabel="count")
```

## 'concave points_mean' and diagnosis:
    * we could see here overlapping between B & M.
    * we could see that the overlapping is only on part of the range.
    * we could see that the min,max,IQR, and the average are different between two plots.
    * we could see outliers.

```python

fig, (ax1,ax2) = plt.subplots(1, 2,figsize = (23,8))
pal = ['orange', 'blue']
sns.kdeplot(ax=ax1, data=df_small, x='concave points_mean', hue="diagnosis",fill=True,palette=pal);
sns.boxplot(ax=ax2, x="diagnosis", y='concave points_mean', data=df_small,palette=pal);
a = ax1.set(title=f"{'concave points_mean'} overlapping malingnant and benign",xlabel=f"{'concave points_mean'}",ylabel="count")
b = ax2.set(title=f"{'concave points_mean'} overlapping malingnant and benign",xlabel=f"{'concave points_mean'}",ylabel="count")
```

## radius_mean vs concave points_mean

### we could see in the plot that there is a small overlapping, but the combination of this two variables gives us pretty good ability to distinguish between B & M


```python
fig, ax = plt.subplots(1, 1,figsize = (10,10))
pal = ['orange','blue']
sns.scatterplot(ax=ax, data=df_small, x="radius_mean", y="concave points_mean", hue="diagnosis",palette=pal);
```
